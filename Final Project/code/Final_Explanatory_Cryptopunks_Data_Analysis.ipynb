# Import Data

```python
# Import the Processed Decentralization Index for AAVE
df = pd.read_csv('https://raw.githubusercontent.com/SunYutongAmber/portfolio/main/Final%20Project/data/final_queried_data..csv')
df.head()
```

# Visualization

## Visualization for Cryptopunks' Price over time

```python
fig1 = make_subplots()
fig1.add_trace(go.Scatter(x=df['timestamp'], y=df['eth'],
              mode='lines', name='price'))
### homework: add comp and aave to the code
fig1.update_layout(title='Scatter Plot for Average price of Cryptopunks over time' ,autosize=True,)
py.iplot(fig1)
```

```python
fig2 = make_subplots()
fig2.add_trace(go.Histogram2dContour(x=df['timestamp'], y=df['eth']))
### homework: add comp and aave to the code
fig2.update_layout(title='Histogram Contour for Average price of Cryptopunks over time' ,autosize=True,)
py.iplot(fig2)
```

# Visualization for Cryptopunks' Return on Investment (ROI) over time

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig3 = make_subplots()
df=df[df['ROI']<1]
fig3.add_trace(go.Scatter(x=df['timestamp'], y=df['ROI'],
              mode='lines', name='price'))
### homework: add comp and aave to the code
fig3.update_layout(title='Scatter Plot for Return on Investment(ROI) of Cryptopunks over time' ,autosize=True,)
fig3.update_xaxes(
    title={'font': {'size': 18}, 'text': 'Date', 'standoff': 10},
    automargin=True)
fig3.update_yaxes(
    title={'font': {'size': 18}, 'text': 'Return on Investment (ETH)', 'standoff': 10},
    automargin=True)
py.iplot(fig3)
```

```python
fig4 = make_subplots()
fig4.add_trace(go.Histogram2dContour(x=df['timestamp'], y=df['ROI']))
### homework: add comp and aave to the code
fig4.update_layout(title='Histogram Contour for Return on Investment(ROI) of Cryptopunks over time' ,autosize=True)
fig4.update_xaxes(
    title={'font': {'size': 18}, 'text': 'Date', 'standoff': 10},
    automargin=True)
fig4.update_yaxes(
    title={'font': {'size': 18}, 'text': 'Return on Investment (ETH)', 'standoff': 10},
    automargin=True)
py.iplot(fig4)
```
